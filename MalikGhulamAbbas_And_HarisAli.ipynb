```python
train['Date of Joining'] = pd.to_datetime(train['Date of Joining']) #Converts the "Date of Joining" column in the "train" DataFrame from its current format
train
```

```python
def get_y_m_d(train): #Define a function named with "get_y_m_d" train datatime as input
    train['Years'] = train['Date of Joining'].dt.year #Creates a new column called "Years" in train DataFrame using dt assign values
    train['Months'] = train['Date of Joining'].dt.month #Creates a new column called "Months" in train DataFrame using dt assign val
    train['Date'] = train['Date of Joining'].dt.day #Creates a new column called "Date" in train DataFrame using dt assign val
    return train 
train = get_y_m_d(train) #Calls the function and assigned the return DataFrame
train.drop(columns =['Date of Joining'],inplace = True) #Drop the columns of "Date of Joining" from train DataFrame
train
```

```python
from sklearn.preprocessing import LabelEncoder #import the labelEncoder from sklearn.preprocessing module
lab_enc = LabelEncoder() #create the instance and assigned it with "lab_enc"
train['Gender']=lab_enc.fit_transform(train['Gender']) #Encodes the "Gender" column in train DataFrame and fit with "lab_enc" variable
train['Company Type']=lab_enc.fit_transform(train['Company Type']) #Encodes the "Company Name" column in train DataFrame and fit with "lab_enc" variable
train['WFH Setup Available']=lab_enc.fit_transform(train['WFH Setup Available']) #Encodes the "WFH Setup Available" column in train DataFrame and fit with "lab_enc" variable
train #show the contents in the train DataFrame
```
